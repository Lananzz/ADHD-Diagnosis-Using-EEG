# Merge all the files together

```python
import os
```

```python
#31 adhd kid
dataset_path = os.listdir('adhd')
label_types = os.listdir('adhd')
label_types.remove(".DS_Store")
print(label_types)  
```

```python
#to concat all the files in one dataframe
frames=[]
c=1
rownum=0
for i in label_types:
    mat_data = loadmat('adhd/'+i)
    variable_data = mat_data[i[:-4]]
    name="df"+str(c)
    name=pd.DataFrame(variable_data)
    #to know the number of row
    m=name.shape
    rownum+=m[0]
    frames+=[name]
    adhd = pd.concat(frames)
    c+=1
print("the number of actual rows: "+str(rownum))
adhd.rename(columns={0:'Fz',1:'Cz',2:'Pz',3:'C3',4:'T3',5:'C4',6:'T4',7:'Fp1',8:'Fp2',9:'F3',10:'F4',11:'F7',12:'F8',13:'P3',14:'P4',15:'T5',16:'T6',17:'O1',18:'O2'},inplace=True)
adhd=pd.DataFrame(adhd)
adhd
```

```python
#add a new column called label
adhd['Label']='adhd'
adhd
```

# will do the same with the normal kids files

```python
#30 normal kid
dataset_path = os.listdir('normal')
label_types = os.listdir('normal')
#label_types.remove(".DS_Store")
print(label_types) 
```

```python
#to concat all the files in one dataframe
frames=[]
c=1
rownum=0
for i in label_types:
    mat_data = loadmat('normal/'+i)
    variable_data = mat_data[i[:-4]]
    name="dfn"+str(c)
    name=pd.DataFrame(variable_data)
    #to know the number of row
    m=name.shape
    rownum+=m[0]
    frames+=[name]
    normal = pd.concat(frames)
    c+=1
print("the number of actual rows: "+str(rownum))
normal.rename(columns={0:'Fz',1:'Cz',2:'Pz',3:'C3',4:'T3',5:'C4',6:'T4',7:'Fp1',8:'Fp2',9:'F3',10:'F4',11:'F7',12:'F8',13:'P3',14:'P4',15:'T5',16:'T6',17:'O1',18:'O2'},inplace=True)
normal=pd.DataFrame(normal)
normal
```

```python
#add a new column called label
normal['Label']='normal'
normal
```

```python
alll=[normal,adhd]
theone= pd.concat(alll)
print("the actual number or rows are: "+str(len(normal)+len(adhd)))
```
